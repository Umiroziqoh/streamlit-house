# Indemepndet Dependent Variabel

```python
dshouse = pd.read_csv('house.csv')
x = dshouse.iloc[:, [0, 1, 2, 3, 4, 5, 6]].values
y = dshouse.iloc[:, 7].values
```

```python
x
```

```python
y
```

# Splint Train Test

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.4,random_state=42)
```

```python
x_test
```

```python
y_test
```

# Model Regresi

```python
from sklearn.linear_model import LinearRegression
regresi =LinearRegression()
```

```python
regresi.fit(x_train,y_train)
```

# Visualisasi Data dan Garis Regresi

```python
independent_column = dshouse.columns[dshouse.columns != 'House_Price']
for column in independent_column :
    plt.figure(figsize=(6, 4))
    sns.regplot(x=dshouse[column],y=dshouse['House_Price'],scatter_kws={'alpha': 0.5},line_kws={"color": "red"})
    plt.xlabel(column)
    plt.ylabel('House_Price')
    plt.title(f'garis regresi {column} vs House Price')
    plt.show()
```

# Testing Model

```python
y_prediksi = regresi.predict(x_test)
print(y_prediksi)
```
